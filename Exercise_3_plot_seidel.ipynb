```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range for x1 and x2
x1 = np.linspace(-2, 2, 400)
x2 = np.linspace(-2, 2, 400)

# Create a meshgrid for x1 and x2
X1, X2 = np.meshgrid(x1, x2)

# Define the equations
equation1 = X1**2 + X2**2 - 1
equation2 = 2*X1**2 - X2 - 1

# Plot the equations
plt.figure(figsize=(8, 6))
plt.contour(X1, X2, equation1, levels=[0], colors='blue', label='eqn1')
plt.contour(X1, X2, equation2, levels=[0], colors='red', label='eqn2')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Intersection of Equations')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
```

```python
import numpy as np

# Coefficient matrix C
C = np.array([[1/4, 1/8],
              [-1/8, -1/4]])

# Constant vector D
D = np.array([[-11/8],
              [-1]])

# Initial guess
X = np.array([[0],
              [0]])

# Maximum number of iterations
max_iterations = 1000

# Tolerance for convergence
tolerance = 1e-6

# Seidel iteration
for _ in range(max_iterations):
    X_new = np.copy(X)
    for i in range(len(X)):
        X_new[i] = (D[i] - np.dot(C[i, :i], X_new[:i]) - np.dot(C[i, i + 1:], X[i + 1:])) / C[i, i]
    if np.allclose(X, X_new, atol=tolerance):
        break
    X = X_new

print("Solution using Seidel method:")
print(X.flatten())
```

```python
import numpy as np

# Define the fixed-point iteration function G(X)
def G(X):
    return np.array([-2*X[1] - 11/2,
                     -2*X[0] + 4])

# Initial guess
X = np.array([0, 0])

# Maximum number of iterations
max_iterations = 1000

# Tolerance for convergence
tolerance = 1e-6

# Initialize iteration count
iterations = 0

# Fixed-point iteration
for _ in range(max_iterations):
    X_new = G(X)
    iterations += 1
    if np.allclose(X, X_new, atol=tolerance):
        break
    X = X_new

print("Solution using fixed-point iteration:")
print(X)
print("Number of iterations:", iterations)
```
