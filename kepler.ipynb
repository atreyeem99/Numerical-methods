```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Data
planets = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn']
ln_P = [-1.418, -0.480, 0.000, 0.631, 2.473, 3.379]
ln_a = [-0.947, -0.322, 0.000, 0.420, 1.648, 2.253]

# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(ln_a, ln_P)

# Exponent
n = slope

# Create a range of ln(a) values for plotting the regression line
ln_a_range = np.linspace(min(ln_a), max(ln_a), 100)
ln_P_fit = slope * ln_a_range + intercept

# Plot
plt.figure(figsize=(8, 6))
plt.plot(ln_a, ln_P, 'bo', label='Data Points')
plt.plot(ln_a_range, ln_P_fit, 'r-', label='Linear Regression')
plt.xlabel('ln(a)')
plt.ylabel('ln(P)')
plt.title('Kepler\'s Third Law - Linear Regression')
plt.legend()
plt.grid(True)
plt.show()

# Print results
print("Exponent (n):", n)
print("R-squared:", r_value**2)
```
