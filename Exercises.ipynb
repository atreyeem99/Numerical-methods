Exercise 3: Write a Python program to find the square-root of 5 by solving the equation x 2 − 5 = 0 using the
bisection method in the interval [1,3] . How many iterations does it take to reach | a − b | < 0.001 and
compare with the number of iterations taken in fixed-point-iteration method to reach | xn+1 − xn | < 0.001.

```python
def bisection_method(func, a, b, tolerance):
    iterations = 0
    while (b - a) / 2 > tolerance:
        c = (a + b) / 2
        if func(c) == 0:
            break
        elif func(c) * func(a) < 0:
            b = c
        else:
            a = c
        iterations += 1

    return (a + b) / 2, iterations

def fixed_point_iteration(func, initial_guess, tolerance):
    iterations = 0
    current_value = initial_guess

    while True:
        next_value = func(current_value)

        if abs(next_value - current_value) < tolerance:
            break

        current_value = next_value
        iterations += 1

    return current_value, iterations

# Define the equation function
def equation(x):
    return x**2 - 5

# Define the fixed-point iteration function for sqrt(5)
def sqrt5_iteration(x):
    return 0.5 * (x + 5 / x)

# Initial interval [1, 3] for bisection method
a_interval = 1
b_interval = 3

# Set tolerance for bisection method
bisection_tolerance = 0.001

# Apply bisection method
bisection_result, bisection_iterations = bisection_method(equation, a_interval, b_interval, bisection_tolerance)

print(f"Bisection Method Result: {bisection_result}")
print(f"Number of Bisection Iterations: {bisection_iterations}")

# Initial guess for fixed-point iteration
initial_guess_fixed_point = 2

# Set tolerance for fixed-point iteration
fixed_point_tolerance = 0.001

# Apply fixed-point iteration method
fixed_point_result, fixed_point_iterations = fixed_point_iteration(sqrt5_iteration, initial_guess_fixed_point, fixed_point_tolerance)

print(f"\nFixed-Point Iteration Result: {fixed_point_result}")
print(f"Number of Fixed-Point Iteration Iterations: {fixed_point_iterations}")
```

Write a Python program to find the square-root of 5 by solving the equation x 2 − 5 = 0 using the Newton-
Raphson method starting with x0 = 1. How many iterations does it take to reach | xn+1 − xn | < 0.001.
Multiple roots: If f(x) has more than one root at x = x* (for example, f(x) = (x − x*)M ), then show that
Newton-Raphson iteration fails to converge. Show that in this case, the modified Newton-Raphson
iteration given below works
xn+1 = xn −
f(xn)f (1)(xn)
[ f (1)(xn)] − f(xn)f (2)(xn)

```python
def f(x):
    return x**2 - 5

def f_prime(x):
    return 2 * x

def newton_raphson_method(func, func_prime, initial_guess, tolerance):
    iterations = 0
    while True:
        next_value = initial_guess - func(initial_guess) / func_prime(initial_guess)
        if abs(next_value - initial_guess) < tolerance:
            break
        initial_guess = next_value
        iterations += 1
    return next_value, iterations

# Initial guess for Newton-Raphson method
initial_guess_newton = 1

# Tolerance for Newton-Raphson method
newton_tolerance = 0.001

# Apply Newton-Raphson method
newton_result, newton_iterations = newton_raphson_method(f, f_prime, initial_guess_newton, newton_tolerance)

print(f"Newton-Raphson Method Result: {newton_result}")
print(f"Number of Newton-Raphson Iterations: {newton_iterations}")
```

Excercise 6: Find example equations to show using fixed-point iteration:
(i) monotonous convergence, (ii) monotonous divergence,
(iii) oscillatory convergence, (iv) oscillatory divergence.
Present your results in the form of cobweb diagrams.

```python
import numpy as np
import matplotlib.pyplot as plt

def fixed_point_iteration(func, x0, n_iterations):
    x_values = [x0]
    y_values = [x0]

    for i in range(1, n_iterations + 1):
        x_values.append(func(x_values[i - 1]))
        y_values.append(x_values[i - 1])

    return x_values, y_values

# Define functions
def func1(x):
    return x / 2 + 2

def func2(x):
    return 2 * x - 3

def func3(x):
    return np.cos(x)

def func4(x):
    return np.sin(x)

# Set parameters
initial_guess = 1
iterations = 10

# Plot cobweb diagrams
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# (i) Monotonous Convergence
x1, y1 = fixed_point_iteration(func1, initial_guess, iterations)
axs[0, 0].plot(x1, y1, 'r.-', label='Iteration')
axs[0, 0].plot(x1, x1, 'b-', label='Identity')
axs[0, 0].set_title('Monotonous Convergence')

# (ii) Monotonous Divergence
x2, y2 = fixed_point_iteration(func2, initial_guess, iterations)
axs[0, 1].plot(x2, y2, 'r.-', label='Iteration')
axs[0, 1].plot(x2, x2, 'b-', label='Identity')
axs[0, 1].set_title('Monotonous Divergence')

# (iii) Oscillatory Convergence
x3, y3 = fixed_point_iteration(func3, initial_guess, iterations)
axs[1, 0].plot(x3, y3, 'r.-', label='Iteration')
axs[1, 0].plot(x3, x3, 'b-', label='Identity')
axs[1, 0].set_title('Oscillatory Convergence')

# (iv) Oscillatory Divergence
x4, y4 = fixed_point_iteration(func4, initial_guess, iterations)
axs[1, 1].plot(x4, y4, 'r.-', label='Iteration')
axs[1, 1].plot(x4, x4, 'b-', label='Identity')
axs[1, 1].set_title('Oscillatory Divergence')

# Set common labels and legend
for ax in axs.flat:
    ax.set(xlabel='x', ylabel='f(x)')
    ax.legend()

plt.tight_layout()
plt.show()
```

Exercise 7: As you have seen in the example code, the computational complexity of
addition of two matrices is (N 2) , and for multiplication it is (N 3) . Write
Python functions to
(1) add two circulant matrices with a computational complexity of (N ) , i.e., a
linear scaling procedure.
(2) multiply two circulant matrices with a computational complexity of (N 2) ,
i.e., a quadratically scaling procedure.
You can figure out the definition of a circulant matrix by printing a square
Toeplitz matrix using the scipy.linalg.toeplitz function and passing
the first row of the matrix as the argument.

```python
import numpy as np
from scipy.linalg import circulant

def add_circulant_matrices(A, B):
    """
    Add two circulant matrices.

    Parameters:
    - A, B: Circulant matrices represented as 1D NumPy arrays (first row).

    Returns:
    - C: Resulting circulant matrix.
    """
    return np.add(A, B)

def multiply_circulant_matrices(A, B):
    """
    Multiply two circulant matrices.

    Parameters:
    - A, B: Circulant matrices represented as 1D NumPy arrays (first row).

    Returns:
    - C: Resulting circulant matrix.
    """
    N = len(A)
    C = np.zeros(N, dtype=complex)

    for k in range(N):
        C[k] = sum(A[i] * B[(k - i) % N] for i in range(N))

    return C

# Example usage:
# Define first rows of two circulant matrices
row_A = np.array([1, 2, 3, 4])
row_B = np.array([5, 6, 7, 8])

# Create circulant matrices
circulant_A = circulant(row_A)
circulant_B = circulant(row_B)

# (1) Add two circulant matrices
result_addition = add_circulant_matrices(row_A, row_B)
print("Addition Result:")
print(result_addition)

# (2) Multiply two circulant matrices
result_multiplication = multiply_circulant_matrices(row_A, row_B)
print("\nMultiplication Result:")
print(result_multiplication)
```

Exercise 7b(6 orig) :Write a Python function for reading a matrix from a file if the elements are listed sequentially, one below
another. Use reshape function to convert a 1D array into a 2D array (matrix).

```python
import numpy as np

def read_matrix_from_file(file_path, rows, columns):
    """
    Read a matrix from a file where elements are listed sequentially.

    Parameters:
    - file_path: Path to the file containing the matrix elements.
    - rows: Number of rows in the matrix.
    - columns: Number of columns in the matrix.

    Returns:
    - matrix: 2D NumPy array representing the matrix.
    """
    try:
        # Read the 1D array from the file
        with open(file_path, 'r') as file:
            content = file.read().split()
            flat_array = np.array([float(x) for x in content])

        # Reshape the 1D array into a 2D array (matrix)
        matrix = np.reshape(flat_array, (rows, columns))

        return matrix

    except Exception as e:
        print(f"Error reading matrix from file: {e}")
        return None

# Example usage:
file_path = 'matrix_data.txt'  # Replace with your file path
num_rows = 3
num_columns = 4

result_matrix = read_matrix_from_file(file_path, num_rows, num_columns)

if result_matrix is not None:
    print("Matrix read from file:")
    print(result_matrix)
```
